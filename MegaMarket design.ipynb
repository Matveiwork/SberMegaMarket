## Дизайн A/B теста: стоит ли внедрять модель по расчету интервала времени доставки


### Бизнес проблема и текущая ситуация: В продукте есть три вида доставки: плановая доставка(длительный период времени доставки),самовывоз(клиент сам забирает заказ),быстрая доставка(доставка за час).Наша задача концентрируется на быстрой доставке,на данный момент определение времени доставки осуществляется по системе указания времени для всех заказов от 20 до 60 минут,никак не ориентироваясь на местоположение клиента,проблема такого метода заключалась в том,что некоторые клиенты получали заказы позже, чем ожидали. К тому же интервал довольно широк, поэтому людям не до конца понятно, в какое время будет доставка. Это создаёт негативный опыт для клиентов, из-за чего они могут выбрать другие сервисы, в которых информация о времени доставки больше соответствует действительности и реже срываются сроки.

### Гипотеза: Если мы внедрим систему более точного расчета времени,то конверсия в покупку увеличится

### Прибыль от решения проблемы: изменение системы расчета времени доставки увеличит конверсию покупки,путем более индидуального расчета времени и у клиента возрастет интерес,так же снизит количество просрочек по времени доставки,что увеличит количество вовремя привезенных заказов и уменьшит количество затрат на решения проблем с задерживающимися заказами,это может увеличить прибыль компании от 3% до 10%

### Описание эксперимента:A/B тест будет проводиться путем сплита пользователей на две группу.Одна группа будет видеть старую систему времени доставки(20-60 минут) -- это будет контрольная группа,вторая группа будет видеть время доставки расчитанная новой моделью в зависимости от удалённости клиента от магазина, времени года, погодных условий, количества активных заказов в магазине и курьеров и так далее.В процессе эксперимента будет измеряться метрики совершения покупки клиентом и выполнения вовремя доставки.Через 7 дней проведения эксперимента метрики будут сравниваться.Если различия в метриках будут больше порога значимости(5%),то эксперимент будет разворачиваться на весь сервис.

### Метрики для анализа эксперимента: количество клиентов совершившых заказ и количество заказов доставленных вовремя.
